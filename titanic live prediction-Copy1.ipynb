```python
def step_gradient(point,learning_rate,m):
    m_slope = np.zeros(8)
    n=len(point)
    for i in range (n):
        x=point[i,:]
        y=y1[i]
        for j in range (len(x)):
            s=[]
            for v in range (len(x)):
                s.append(x[v]*m[v])
            h=1/(1+math.exp(-1*sum(s)))
            m_slope[j]+=(-1/n)*(y-h)*x[j]
            
    new_m=m-learning_rate*m_slope
   
    return new_m
```

```python
def gd(point,learning_rate,num_iteration):
    m = np.zeros(8)
   
    for i in range (num_iteration):
        m=step_gradient(point,learning_rate,m)
        print(i,"cost",cost(point,m))
    return m

```

```python
def cost(point, m):
    total_cost=0
    n=len(point)
    for i in range(n):
        x = point[i,:]
        y = y1[i]
        b=[]
        for j in range(len(m)):
            b.append (m[j]*x[j])
        h = 1 / (1+math.exp(-1*sum(b))) 
        #if sum(b) < 0:
            #h= 1 - 1/(1 + math.exp(sum(b)))
        #else:
            #h= 1/(1 + math.exp(-sum(b)))
        total_cost += (-1/n)* (y*(math.log(h))+((1-y)*(math.log(1-h))))
        #print(total_cost)
    return total_cost
```

```python
def predict(x3,m):
    y3=[]
    n=len(x3)
    for i in range (n):
        s=[]
        x=x3[i,:]
        for j in range (len(m)):
            s.append(m[j]*x[j])
        h=1/(1+math.exp(-1*sum(s)))
        print(h)
        if h>0.5:
            y3.append(1)
        elif h<=0.5:
            y3.append(0)
    return y3
```
